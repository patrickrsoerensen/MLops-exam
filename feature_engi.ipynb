```python
# Splitting categories and genres
df['Categories'] = df['Categories'].str.split(',')
df['Genres'] = df['Genres'].str.split(',')
```

```python
df
```

```python
# Check for NaN values in the entire DataFrame
nan_values = df.isna().sum().sum()
print("Number of NaN values in the DataFrame:", nan_values)
```

```python
# One-hot encoding categories and genres
categories_dummies = pd.get_dummies(df['Categories'].apply(pd.Series).stack()).sum(level=0)
genres_dummies = pd.get_dummies(df['Genres'].apply(pd.Series).stack()).sum(level=0)
```

```python
# Check for NaN values in the entire DataFrame
nan_values = df.isna().sum().sum()
print("Number of NaN values in the DataFrame:", nan_values)
```

```python
# Concatenating one-hot encoded columns with original DataFrame
df = pd.concat([df, categories_dummies, genres_dummies], axis=1)
```

```python
# Check for NaN values in the entire DataFrame
nan_values = df.isna().sum().sum()
print("Number of NaN values in the DataFrame:", nan_values)
```

```python
# Dropping original 'Categories' and 'Genres' columns
df.drop(columns=['Categories', 'Genres'], inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Process textual column (About the Game) with adjusted parameters
tfidf_vectorizer = TfidfVectorizer(max_features=1000, max_df=0.8)
about_game_features = tfidf_vectorizer.fit_transform(df['About the game'])
about_game_df = pd.DataFrame(about_game_features.toarray(), columns=tfidf_vectorizer.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Process game title using TF-IDF with adjusted parameters
tfidf_vectorizer_title = TfidfVectorizer(max_features=1000, max_df=0.8)
title_features = tfidf_vectorizer_title.fit_transform(df['Name'])
title_df = pd.DataFrame(title_features.toarray(), columns=tfidf_vectorizer_title.get_feature_names_out())
```

```python
df.reset_index(drop=True, inplace=True)
about_game_df.reset_index(drop=True, inplace=True)
title_df.reset_index(drop=True, inplace=True)
```

```python
# Concatenate processed dataframes
df = pd.concat([df, about_game_df, title_df], axis=1)
```

```python
# Drop original 'About the game' and 'Name' columns
df.drop(columns=['About the game', 'Name'], inplace=True)
```
