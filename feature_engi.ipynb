```python
# Function to preprocess text using spaCy
def preprocess_text(text):
    # Parse the text with spaCy
    doc = nlp(text)
    # Lemmatize the tokens and remove stopwords
    tokens = [token.lemma_ for token in doc if not token.is_stop]
    # Join the tokens back into a string
    preprocessed_text = " ".join(tokens)
    return preprocessed_text

# Assuming df is your DataFrame with "Name" and "About the game" columns
# Apply the preprocess_text function to the "Name" column
df['Name'] = df['Name'].apply(preprocess_text)

# Apply the preprocess_text function to the "About the game" column
df['About the game'] = df['About the game'].apply(preprocess_text)

# Display the preprocessed DataFrame
print(df)
```
